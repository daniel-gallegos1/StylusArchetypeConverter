## Testing snapping a stroke based on 2 anchor points:
Given a stroke with a starting and ending spot, it may be useful to adjust the bounding box so that the start/end point is adjusted to two other points (e.g. to connect to 2 other stroke start/end points.

Consider a given stroke with the bound defined by two points $p_0$ and $p_1$ with the stroke start beginning at $s$ and ending at $e$. Now suppose we want to stretch and translate the strokes bounding box so the stroke starting and ending points are mapped to $a$ and $b$, respectively.

We need to figure out a $n_0$ and $n_1$ to be the new points defining the bounding box. Let $n_0 = p_0 + d_0$ and $n_1 = p_1 + d_1$

</br>

Solution:

Construct $f_s = \frac{\left(s-p_{0}\right)}{p_{1}-p_{0}}$ which is the fractional distance $s$ is within the bounding box. Construct $f_e$ similarly.

We know $s = (1-f_s)p_0 + f_s p_1$, similarly for $e$. Likewise $a = (1-f_s)(p_0 + d_0) + f_s(p_1 + d_1)$, similarly for $b$.

After some manipulation we find $a-s = (1-f_s)d_0 + f_s d_1$ and $(a-s)-f_s d_1)/(1-f_s) = d_0 = (b-e)-f_e d_1)/(1-f_e)$.

Yada yada yada, more algebra we get $d_1 = \frac{(1-f_e)(s-a)+(1-f_s)(b-e)}{f_e-f_s}$ Note this formula is undefined if we have two points which have the same $y$ or $x$ coordinates.

Also $d_0=\frac{(a-s)-f_s d_1}{1-f_s} = \frac{(b-e)-f_e d_1}{1-f_e}$.

```python
import matplotlib.pyplot as plt
import bezier

def graphFits(fits, resol=25):
    fig, ax = plt.subplots(figsize=(16,16))
    ax.invert_yaxis()

    evalPoints = np.linspace(0.0, 1.0, resol) #determines how accurate bezier curves are drawn
    for fit in fits: #plot bounding box and bezier curves
        arr2 = np.ones(2) # just [1,1], makes these next lines a bit more readable
        ax.plot( fit.x[0]*arr2, fit.y, linestyle="dashed", color="blue", linewidth=1) #plot bounding box
        ax.plot( fit.x[1]*arr2, fit.y, linestyle="dashed", color="blue", linewidth=1)
        ax.plot( fit.x, fit.y[0]*arr2, linestyle="dashed", color="blue", linewidth=1)
        ax.plot( fit.x, fit.y[1]*arr2, linestyle="dashed", color="blue", linewidth=1)
        
        for curve in sf.mapStroke(fit.data.arial, fit.x, fit.y): #control points, 
            nodes = np.array(curve).transpose()
            curve = bezier.Curve(nodes, degree=len(curve)-1)
            curvePoints = np.array(curve.evaluate_multi(evalPoints).transpose().tolist()) #[[x,y]...]
            ax.plot(curvePoints[:,0], curvePoints[:,1], color="black", linewidth=3)
    return fig, ax
```

```python
def graphFits(fits, resol=25):
    fig, ax = plt.subplots(figsize=(16,16))
    ax.invert_yaxis()

    evalPoints = np.linspace(0.0, 1.0, resol) #determines how accurate bezier curves are drawn
    for fit in fits: #plot bounding box and bezier curves
        arr2 = np.ones(2) # just [1,1], makes these next lines a bit more readable
        ax.plot( fit.x[0]*arr2, fit.y, linestyle="dashed", color="blue", linewidth=1) #plot bounding box
        ax.plot( fit.x[1]*arr2, fit.y, linestyle="dashed", color="blue", linewidth=1)
        ax.plot( fit.x, fit.y[0]*arr2, linestyle="dashed", color="blue", linewidth=1)
        ax.plot( fit.x, fit.y[1]*arr2, linestyle="dashed", color="blue", linewidth=1)
        
        for curve in sf.mapStroke(fit.data.arial, fit.x, fit.y): #control points, 
            nodes = np.array(curve).transpose()
            curve = bezier.Curve(nodes, degree=len(curve)-1)
            curvePoints = np.array(curve.evaluate_multi(evalPoints).transpose().tolist()) #[[x,y]...]
            ax.plot(curvePoints[:,0], curvePoints[:,1], color="black", linewidth=3)
    return fig, ax

def mapCoord(fit, coord):
    temp = np.zeros((2))
    temp[0] = (1-coord[0])*fit.x[0] + coord[0]*fit.x[1]
    temp[1] = (1-coord[1])*fit.y[0] + coord[1]*fit.y[1]
    return temp

def mapStroke(strokeData, xR, yR):
    strokeData = np.array(strokeData)
    strokeData[:,:,0] = xR[0] + (strokeData[:,:,0]*(xR[1]-xR[0]))
    strokeData[:,:,1] = yR[0] + (strokeData[:,:,1]*(yR[1]-yR[0]))
    return strokeData

def adjustFit(fit, newS, newE, tol=.001): #note newS/newE coord order is [y,x]
    p0 = np.array([fit.x[0], fit.y[0]])
    p1 = np.array([fit.x[1], fit.y[1]])
    
    fs = np.array(fit.data.arial[0][0])
    fe = np.array(fit.data.arial[-1][-1])
    
    s = (1-fs)*p0 + fs*p1
    e = (1-fe)*p0 + fe*p1
    
    d0 = np.zeros((2))
    d1 = np.zeros((2))
    
    if(abs(fs[0]-fe[0]) >= tol): #check for divide by zero
        d1[0] = ((1-fe[0])*(s[0] - newS[1]) + (1-fs[0])*(newE[1] - e[0]))/(fe[0]-fs[0])
    if(abs(fs[1]-fe[1]) >= tol): #check for divide by zero
        d1[1] = ((1-fe[1])*(s[1] - newS[0]) + (1-fs[1])*(newE[0] - e[1]))/(fe[1]-fs[1])
    
    if(abs(1-fs[0]) >= tol):
        d0[0] = ((newS[1] - s[0]) - fs[0]*d1[0]) / (1-fs[0])
    elif(abs(1-fe[0]) >= tol):
        d0[0] = ((newE[1] - e[0]) - fe[0]*d1[0]) / (1-fe[0])
        
    if(abs(1-fs[1]) >= tol):
        d0[1] = ((newS[0] - s[1]) - fs[1]*d1[1]) / (1-fs[1])
    elif(abs(1-fe[1]) >= tol):
        d0[1] = ((newE[0] - e[1]) - fe[1]*d1[1]) / (1-fe[1])
    #d1 = ((1-fe)*(newS - s) + (1-fs)*(newE - e))/(fe-fs)
    #d0 = ((newS - s) - fs*d1) / (1-fs)
    newP0 = p0+d0
    newP1 = p1+d1
    return sf.Fit(fit.name, fit.data, [newP0[0], newP1[0]], [newP0[1], newP1[1]])
    #if(not np.any(np.isclose(fs, fe, atol=.001))): #x or y coordinates don't match
    #    return fit #return as is for now
        #find closest point and only map to that one
    
```

## Determining anchor points:
Each stroke has 2 endpoints

__Strokes ends can only be of a couple different types:__

Single end point: | (a single stroke end such that it intersects with no other strokes)

Corner end point: L (two stroke end points joined at some angle, normally 90 degrees)

T end points: T, (one end point joined along the line of another stroke)

An X intersection could technically include an end point, but this seems to be rare and normally doesn't involve any end points.

<br>

For each type we can (1) determine the type of end point by the local corners (there are two corner types normal (3/4 whitespace), or inverted (1/4 whitespace)), (2) determine the end point based on the stroke end type:

<br>

__(1) determine end type: (all these local corners should be close)__

Single: 2 normal corners at with a ~90 degree angle

Corner: 1 normal corner and 1 inverted corner with the same angle

T: Two inverted corners with complementary angles

<br>


__(2) Determine end point position:__

Single: Midpoint of the two corners

Corner: Midpoint of the two corners

T: The intersection of the vectors of the two angles, this actually works slightly poorly, I've swapped the method, to ~sqrt(2)/2 from each of the corners

```python
import numpy as np
import FormalImage as fi
import matplotlib.pyplot as plt
import Anchors as an

img = fi.renderChar("\u4f4d", size=1000, pad=.1, show=False);

anchConv = an.AnchorConverter()

anchConv.DiffBorderDist=10
anchConv.diffThresh=.2

diffImg = anchConv._getDiff(img, borderDiff=anchConv.DiffBorderDist)
anchors, corners = anchConv.get(img, getCorners=True)
```

```python
fig, ax = plt.subplots(figsize=(16,16))

ax.imshow(img)
ax.set_xlim([0,1000-1])
ax.set_ylim([0,1000-1])

for a in anchors:
    ax.plot(a[1], a[0], ".", color="blue", markersize=20)
for c in corners:
    ax.plot(c.coord[1], c.coord[0], ".", color="green", markersize=15)
ax.invert_yaxis()
```

## Now map hanzi positioning to best fit with the arial anchors
We ideally want to map strokes by moving start/ends to the anchor points.
We will define an error metric which is a combination of: preserving aspect ratio ((and avoiding flipping), this aspect ratio doesn't matter for horizontal/vertical strokes), distance from original position, and how well the mapped stroke fits onto the character

If a certain minimum can't reached for a stroke, we won't move it.

```python
import numpy as np
#back end libraries for loading/saving/modifying our data
import StrokeFit as sf
from FitDef import FitData
import FitDef as fd
from StrokeDef import StrokeType #needed for loading object properly
import StrokeDef as sd
import matplotlib.pyplot as plt
import bezier

fitData = fd.loadFits("\u4eb2")

fig, ax = graphFits(fitData.fits)

ax.imshow(img, cmap="YlOrRd")
ax.set_xlim([0,1000-1])
ax.set_ylim([0,1000-1])

for a in anchors:
    ax.plot(a[1], a[0], ".", color="blue", markersize=20)
ax.invert_yaxis()
```

```python
#image should be the score of a curve point landing on that position, (positive is good)
#samples is samples per curve
def getMapError(oldFit, newFit, img, samples = 10, mults=[1,1,1]):
    error = 0
    
    sData = oldFit.data.arial
    orthFit = (sData[0][0][0] == sData[-1][-1][0]) or (sData[0][0][1] == sData[-1][-1][1]) #should be horizontal/vertical 
    
    #aspect ratio
    perimeter = np.diff(newFit.x)[0] + np.diff(newFit.y)[0] * 2
    if(perimeter < .25*len(img)): #stroke perimeter is too small
        error = 10e10 #arbitrary large error
        return error
    
    if(np.diff(newFit.x)[0] < len(img)*.01 or np.diff(newFit.y)[0] < len(img)*.01): #some stroke dimension is too small
        error = 10e10 #arbitrary large error
        return error
    
    if(not orthFit and newFit.x[0] > newFit.x[1] or newFit.y[0] > newFit.y[1]): #stroke is flipped, not allowed, unless flipping doesn't do anything
        error = 10e10 #arbitrary large error
        return error
    
    if(not orthFit): #if a stroke is vertical or horizontal we don't care about aspect ratio
        if(newFit.y[0] == newFit.y[1] or newFit.x[0] == newFit.x[1]): #undefined aspect ratio
            error = 10e10 #arbitrary large error
            return error
        oldAspect = np.diff(oldFit.y)[0]/np.diff(oldFit.x)[0]
        newAspect = np.diff(newFit.y)[0]/np.diff(newFit.x)[0]
        error += mults[0] * ( abs(oldAspect - newAspect) + abs(1/oldAspect - 1/newAspect)) #e.g 1:1 to 1:2 would be 1/2 + 1 = 1.5 additional error
        #print( mults[0] * ( abs(oldAspect - newAspect) + abs(1/oldAspect - 1/newAspect)) )
        
    #overall distance change
    imageSize = len(img)
    boundChange = abs(oldFit.x[0] - newFit.x[0]) + abs(oldFit.y[0] - newFit.y[0]) + abs(oldFit.x[1] - newFit.x[1]) + abs(oldFit.y[1] - newFit.y[1])
    
    #print(mults[1] * (boundChange/4) / imageSize)
    error += mults[1] * (boundChange/4) / imageSize #average bound size in percent of image size
    
    #fit error
    fitScore = 0 #positive is good
    evalPoints = np.linspace(0.0, 1.0, samples) #start, end, resol, parametric t values
    
    sData = mapStroke(sData, newFit.x, newFit.y)
    for curve in sData:
        nodes = np.array(curve).transpose()
        curve = bezier.Curve(nodes, degree=len(curve)-1)
        curvePoints = np.array(curve.evaluate_multi(evalPoints).transpose().tolist()) #[[x,y]...]
        for samplePoint in curvePoints:
            y = int(samplePoint[1])
            x = int(samplePoint[0])
            fitScore += img[y,x]
    fitScore = fitScore / (len(sData)*samples) #average of all samples taken
    error -= mults[2] * fitScore
    #print(mults[2] * fitScore)
    return error

def verifyAnchors(fit, a1, a2): #some strokes can't be mapped properly, verify points are near an anchor
    fit2 = mapStroke(fit.data.arial, fit.x, fit.y)
    beginCoord = fit2[0][0]
    endCoord = fit2[-1][-1]
    beginCoord = np.array([beginCoord[1], beginCoord[0]]) #x,y format
    endCoord = np.array([endCoord[1], endCoord[0]]) #x,y format
    
    if(np.sum(abs(beginCoord - a1)) > 10 and np.sum(abs(beginCoord - a2)) > 10): #no matching anchor point
        return False
    if(np.sum(abs(endCoord - a1)) > 10 and np.sum(abs(endCoord - a2)) > 10): #no matching anchor point
        return False
    return True
```

```python
imgError = img.copy()
imgError[img==0] = -1 #penalize whitespace
```

```python
fitDataCopy = FitData()
fitDataCopy.set(fitData.character, [], fitData.size)

minError = 1 #5 is what we need to map the stroke
    
for origFit in fitData.fits:
    bestFit = None
    bestError = 10e10
    for a1 in anchors:
        for a2 in anchors:
            if(np.all(a1 == a2)):
                continue
            newFit = adjustFit(origFit, a1, a2)
            if(not verifyAnchors(newFit,a1,a2)): #map didn't complete
                continue
            error = getMapError(origFit, newFit, imgError, mults=[1,5,5]) #errors: aspect ratio, move distance, fit match
            if(error < bestError):
                bestFit = newFit
                bestError = error
    if(bestError <= minError):
        newFitType = sf.Fit(bestFit.name, bestFit.data, bestFit.x, bestFit.y)
        print("swap")
    else: #use original fit
        newFitType = sf.Fit(origFit.name, origFit.data, origFit.x, origFit.y)
    fitDataCopy.fits.append(newFitType)

fig, ax = graphFits(fitDataCopy.fits)

ax.imshow(img, cmap="YlOrRd")
ax.set_xlim([0,1000-1])
ax.set_ylim([0,1000-1])

for a in anchors:
    ax.plot(a[1], a[0], ".", color="blue", markersize=20)
ax.invert_yaxis()
```
