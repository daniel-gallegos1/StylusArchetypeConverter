```python
import os
import numpy as np

files = os.listdir("..\\Data\\Hanzi Data")
```

```python
chars = []
codes = []
nums = [] #unicode number of each char (decimal)

for fileName in files:
    chars.append(fileName[0])
    codes.append(hex(ord(fileName[0])))
```

```python
lines = []
for i in range(len(chars)):
    lines.append((str(codes[i])[2:] + " " + chars[i]))

for line in lines:
    print(line)
```
